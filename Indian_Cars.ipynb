The average price has consistently **increased year over year**, with a significant jump in 2022 and 2024. There are no periods of significant price decline, indicating a general upward trajectory.

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
plt.figure(figsize=(10,6))
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(np.arange(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.show()
```

**Positive Correlations:**

* **make_year and price:** There's a strong positive correlation between the year of manufacture and the price of the car. This suggests that newer cars tend to have higher prices.
* **engine_capacity(CC) and price:** A moderately strong positive correlation exists between engine capacity and price, indicating that cars with larger engines generally have higher prices.
* **engine_capacity(CC) and make_year:** There's a moderate positive correlation between the year of manufacture and engine capacity, suggesting that newer cars tend to have larger engines.

**Negative Correlation:**

* **km_driven and price:** A moderately strong negative correlation is observed between kilometers driven and price. This indicates that cars with higher mileage tend to have lower prices.

**Weak or No Correlation:**

* **overall_cost and other variables:** The overall_cost variable shows weak or no correlation with the other variables in the dataset. This might suggest that the overall cost is influenced by factors not captured in these variables.

The diagonal elements represent the correlation of a variable with itself, which is always 1.

# **Data Cleaning**

```python
# Fill missing values without inplace=True
df['engine_capacity(CC)'] = df['engine_capacity(CC)'].fillna(df['engine_capacity(CC)'].median())  # 118 entries are missing
df['reg_year'] = df['reg_year'].fillna(df['make_year'])  # 1 entry is missing
df['km_driven'] = df['km_driven'].fillna(df['km_driven'].median())  # 1 entry is missing
df['fuel_type'] = df['fuel_type'].fillna(df['fuel_type'].mode()[0])
df['ownership'] = df['ownership'].fillna(df['ownership'].mode()[0])
df['has_insurance'] = df['has_insurance'].fillna(df['has_insurance'].mode()[0])
df['spare_key'] = df['spare_key'].fillna(df['spare_key'].mode()[0])
df['transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])

# Drop rows with missing values in specific columns
df = df.dropna(subset=['brand', 'model', 'make_year', 'overall_cost', 'reg_number', 'title'])
```

```python
print(df.isnull().sum())

print(df.describe())
```

We have cleaned the data and removed all null entries. The outliers in price has also been removed. Prices range from **Rs.1,63,000 to Rs.24,87,000**. Average price and median becoming 6 lakhs.
